```python
#import a few packages

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Goal
- overlay two normal distributions with different mean and standard deviations
- add legend, xlabel, ylabel, axis ticks, title and save the image in pdf/png format

We'll be producing many plots in the course. It is a good practice to have a common look across the plots. So we will define some global plot settings. We can overwrite them anytime while plotting. 

```python
# Custom Settings

# -- Figsize -- #
mpl.rcParams['figure.figsize']=(8,6)

# --  Axis -- #
mpl.rcParams["axes.linewidth"]  = 2.0
mpl.rcParams["xtick.major.size"]  = 10
mpl.rcParams["xtick.minor.size"]  = 5
mpl.rcParams["ytick.major.size"]  = 10
mpl.rcParams["ytick.minor.size"]  = 5
mpl.rcParams["xtick.direction"]  = "in"
mpl.rcParams["ytick.direction"]  = "in"
mpl.rcParams['axes.grid'] = True
#mpl.rcParams['axes.grid.axis'] = 'y'
mpl.rcParams['grid.color'] = 'grey'
mpl.rcParams['grid.linewidth'] = 0.2

mpl.rcParams["axes.labelsize"]= 20

# -- Font -- #
mpl.rcParams['font.size']=16
mpl.rcParams['font.family']='serif'

# -- Legend -- #
mpl.rcParams['legend.fontsize']=18
mpl.rcParams['legend.frameon']=False

# -- Saving Options --
mpl.rcParams['savefig.bbox'] = 'tight'
mpl.rcParams['savefig.dpi'] = 250
mpl.rcParams['savefig.transparent'] = False
mpl.rcParams['savefig.facecolor'] = 'white'

#plt.rc("text", usetex=True)
```

```python
# Generate two normal distribution
dist1=np.random.normal(0.1,0.6,10000)
dist2=np.random.normal(0.9,0.9,10000)
```

```python
fig, ax = plt.subplots()
ax.hist(dist1, bins=50, lw=2, alpha=0.5,label='dist1')
ax.hist(dist2, bins=50, lw=2, alpha=0.5,label='dist2')
ax.set_xlabel('dist')
ax.set_ylabel('counts')
ax.legend(loc=0)
ax.set_title('Matplotlib basic Normal Scale')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(dist1, bins=50, lw=2, alpha=0.5,label='dist1')
ax.hist(dist2, bins=50, lw=2, alpha=0.5,label='dist2')
ax.set_xlabel('dist')
ax.set_ylabel('counts')
ax.legend(loc=0)
ax.set_title('Matplotlib basic log scale')
plt.yscale('log')
plt.savefig('guassian_logscale.png')

plt.show()
```

## Subplots ( 4x4 grid)

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))
ax = ax.flatten()

##Legend size may look bad, so let's override it
plt.rcParams['legend.fontsize']=15

ax[0].hist(dist1, bins=50, lw=2, alpha=0.5,label='dist1')
ax[0].hist(dist2, bins=50, lw=2, alpha=0.5,label='dist2')
ax[0].set_title('basic')
ax[0].legend(loc='upper left')

ax[1].hist(dist1, bins=50, lw=2, alpha=0.5,label='dist1',density=True)
ax[1].hist(dist2, bins=50, lw=2, alpha=0.5,label='dist2',density=True)
ax[1].set_title('normalized')
ax[1].legend(loc='upper right')

ax[2].hist(dist1, bins=50, lw=2, alpha=0.5,label='dist1',log=True)
ax[2].hist(dist2, bins=50, lw=2, alpha=0.5,label='dist2',log=True)
ax[2].set_title('logscale')
ax[2].legend(loc='lower right')

ax[3].hist(dist1, bins=50, lw=3, histtype='step',alpha=0.5,label='dist1',log=True,density=True)
ax[3].hist(dist2, bins=50, lw=3, histtype='step',alpha=0.5,label='dist2',log=True,density=True)
ax[3].set_title('cosmetic changes')
ax[3].legend(loc='lower center')

plt.show()
```
